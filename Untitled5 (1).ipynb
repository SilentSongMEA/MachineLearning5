```python
from sklearn import datasets
iris=datasets.load_iris()
```

```python
sepal_data=iris.data[:,:2]
petal_data=iris.data[:,2:4]
```

```python
from sklearn.cluster import KMeans
km1=KMeans(n_clusters=3, n_init=3)
km1.fit(sepal_data)
km2=KMeans(n_clusters=3,n_init=3)
km2.fit(petal_data)
```

```python
centroid_sepals=km1.cluster_centers_
labels_sepals=km1.labels_
centroid_petals=km2.cluster_centers_
labels_petals=km2.labels_
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
fig=plt.figure(figsize=(10,10))
ax1=fig.add_subplot(221)
ax2=fig.add_subplot(222,projection="3d")
ax3=fig.add_subplot(223)
ax4=fig.add_subplot(224,projection="3d")
ax1.scatter(sepal_data[:,0],sepal_data[:,1],c=labels_sepals,s=50)
ax1.scatter(centroid_sepals[:,0],centroid_sepals[:,1],c="red",s=100)

ax2.scatter(sepal_data[:,0],sepal_data[:,1],c=labels_sepals,s=50)
ax2.scatter(centroid_sepals[:,0],centroid_sepals[:,1],c="red",s=100)

ax3.scatter(petal_data[:,0],petal_data[:,1],c=labels_petals,s=50)
ax3.scatter(centroid_petals[:,0],centroid_petals[:,1],c="red",s=100)

ax4.scatter(petal_data[:,0],petal_data[:,1],c=labels_petals,s=50)
ax4.scatter(centroid_petals[:,0],centroid_petals[:,1],c="red",s=100)
ax1.set(xlabel=iris.feature_names[0],ylabel=iris.feature_names[1])
ax3.set(xlabel=iris.feature_names[2],ylabel=iris.feature_names[3])
ax1.set_facecolor("pink")
ax2.set_facecolor("pink")
ax3.set_facecolor("pink")
ax4.set_facecolor("pink")
plt.show()
```
