```python
from sklearn.cluster import MeanShift
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
```

```python
x,y=make_blobs(n_samples=100,centers=3,random_state=42,cluster_std=0.9)
```

```python
mean_shift=MeanShift()
mean_shift.fit(x)
```

```python
plt.scatter(x[:,0],x[:,1],c=mean_shift.labels_,cmap='viridis',marker='o')
plt.scatter(mean_shift.cluster_centers_[:,0],mean_shift.cluster_centers_[:,1],
           c='red',marker='x',s=200,linewidths=3,label='Cluster Centers')
plt.title('Mean Shift Clustering')
plt.legend()
plt.show()
```

```python
from sklearn.cluster import MeanShift,KMeans
```

```python
#create synthetic data
x,y=make_blobs(n_samples=300,centers=3,random_state=42)
```

```python
mean_shift=MeanShift()
mean_shift.fit(x)
```

```python
kmeans=KMeans(n_clusters=3,n_init=4)
kmeans.fit(x)
```

```python
plt.figure(figsize=(12,5))
plt.subplot(1,2,1)
plt.scatter(x[:,0],x[:,1],c=mean_shift.labels_,cmap='viridis',marker='o')
plt.scatter(mean_shift.cluster_centers_[:,0],mean_shift.cluster_centers_[:,1],
           c='red',marker='x',s=200,linewidths=3,label='Cluster Centers')
plt.title('Mean Shift Clustering')

plt.subplot(1,2,2)
plt.scatter(x[:,0],x[:,1],c=kmeans.labels_,cmap='viridis',marker='o')
plt.scatter(mean_shift.cluster_centers_[:,0],mean_shift.cluster_centers_[:,1],
           c='red',marker='x',s=200,linewidths=3,label='Cluster Centers')
plt.title('K-Mean Clustering')
plt.show()
```
